# **Pandas**

DataFrame = Excel Spreadsheet
Series = Single Column
Index = row labels



```
import pandas as pd

data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "salary": [70000, 80000, 90000]
}

df = pd.DataFrame(data)

print(df)
```
